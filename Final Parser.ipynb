```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import html2text
import csv
import pandas as pd
import numpy as np
import socket
```

```python
def delit(string):
    string = string.replace(":","")
    string = string.replace("'","")
    string = string.replace('"',"")
    string = string[4:]
    string = string[:-1]
    return string
```

```python
def get_reviews(number_quest):
    quest_link = f"https://otvet.mail.ru/{number_quest}"
    inner_html_code = str(urlopen(quest_link).read(),'utf-8') #read html
    inner_soup = BeautifulSoup(inner_html_code, "html.parser")
    inner_soup = inner_soup.find('script', {'type':'application/ld+json'})
    text = str(inner_soup.get_text()) # found block we convert to string, next step it's just work with string
    reviews_answers = list(re.findall(r'text\":\"[\\а-яА-ЯёЁ \d()\,\-!\."a-zA-Z\/?:]*,',text)) #first element - quest, second - review, third - best answer, another - answers
    sort_reviews_answers = list(map(delit, reviews_answers)) #del trash
    sort_reviews_answers.pop(0)
    quest = str(list(re.findall(r'name":\"[а-яА-Я \d()?!]*\"\,\"text\":',text))) #found need info (quest, reviews etc)
    quest = list(re.findall(r'[а-яА-Я]{1,}',quest)) # to list for sort info
    final_quest =' '.join(quest)
    try:
        review = sort_reviews_answers[0]
    except: 
        review = '0' 
    try:
        best_answer = sort_reviews_answers[1] # in mail the first is the best
    except:
        best_answer = '0'
    another = sort_reviews_answers[2:]
    final_answers = [final_quest, review,best_answer,another, quest_link]
    return final_answers
```

```python
file = open("/home/dimon/Рабочий стол/Текстовый файл….txt", "rb")
need_text = str(file.read())
all_quest = list(re.findall(r'question\/[\d]*',need_text))
unique_quest = list(set((all_quest)))
del unique_quest[318]
del unique_quest[538]
del unique_quest[871]
```

```python

header = ['Quest','Review','Popular','Another','Questslink']
with open('Mailreview.csv', 'w') as outcsv:
    writer = csv.writer(outcsv)
    writer.writerow(header)
    
with open('Mailreview.csv', 'a', encoding='UTF8') as csv_file:
    writer = csv.writer(csv_file)
    for i in unique_quest:
            try:
                writer.writerow(get_reviews(i))
            except:
                write.writewrow(['error','error','error',['error'],i])
            
```
